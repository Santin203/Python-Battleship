```python
# Import libraries
import random
from random import randint
import numpy as np
import pandas as pd
from IPython.display import display, HTML

#Variables
hit = '*'
ai_hit='*'
miss = 'X'
ai_miss='X'
ship = '<>'
ai_ship='<>'
rowcol = [0, 1, 2, 3, 4, 5, 6]
user_hits_num=0
ai_hits_num=0
dif=[1,2]
level=0
play=0
shot_near=0


#Creating boards
user_board = np.zeros((5,5), dtype="object")
user_board2 = np.zeros((7,7), dtype="object")
shot_board = np.zeros((5,5), dtype="object")
shot_board2 = np.zeros((7,7), dtype="object")
ai_board = np.zeros((5,5), dtype="object")
ai_board2 = np.zeros((7,7), dtype="object")


#Functions

#Function to display boards side by side
#Function taked from https://stackoverflow.com/questions/38783027/jupyter-notebook-display-two-pandas-tables-side-by-side
def display_side_by_side(dfs:list, captions:list):
    """Display tables side by side to save vertical space
    Input:
        dfs: list of pandas.DataFrame
        captions: list of table captions
    """
    output = ""
    combined = dict(zip(captions, dfs))
    for caption, df in combined.items():
        output += df.style.set_table_attributes("style='display:inline'").set_caption(caption)._repr_html_()
        output += "\xa0\xa0\xa0"
    display(HTML(output))



#function to place 5 ships on the board
def ship_loc():
    for i in range(5):
        #placing random ships
        user_ship_row = randint(0, 4)
        user_ship_col = randint(0, 4)
        ai_ship_row = randint(0, 4)
        ai_ship_col = randint(0, 4)
        
        if user_board[user_ship_row,user_ship_col]==ship:
            user_ship_row = randint(0, 4)
            user_ship_col = randint(0, 4)
            user_board[user_ship_row,user_ship_col]=ship
            
        else:
            user_board[user_ship_row,user_ship_col]=ship
        
        if ai_board[ai_ship_row,ai_ship_col]==ai_ship:
            ai_ship_row = randint(0, 4)
            ai_ship_col = randint(0, 4)
            ai_board[ai_ship_row,ai_ship_col]=ai_ship
        else:
            ai_board[ai_ship_row,ai_ship_col]=ai_ship
            
            
#function to place 5 ships of size 2 on the board            
def ship_loc2():
    for i in range(5):
        orientation= randint(0, 1)
        
        if orientation == 1:
            
            #placing random ships
            user_ship_row = randint(0, 6)
            user_ship_col = randint(0, 6)
            ai_ship_row = randint(0, 6)
            ai_ship_col = randint(0, 6)

            
            if user_ship_col == 6:
                if user_board2[user_ship_row,user_ship_col]=='<' or user_board2[user_ship_row,user_ship_col]=='>' or user_board2[user_ship_row,user_ship_col]=='^' or user_board2[user_ship_row,user_ship_col]=='v':
                        
                    user_ship_row = randint(0, 6)
                    user_ship_col = randint(0, 6)
                    user_board2[user_ship_row,user_ship_col-1]='<'
                    user_board2[user_ship_row,user_ship_col]='>'
                else:
                    user_board2[user_ship_row,user_ship_col-1]='<'
                    user_board2[user_ship_row,user_ship_col]='>'
                
            elif user_ship_col == 0:
                if user_board2[user_ship_row,user_ship_col]=='<' or user_board2[user_ship_row,user_ship_col]=='>' or user_board2[user_ship_row,user_ship_col]=='^' or user_board2[user_ship_row,user_ship_col]=='v':  
                    user_ship_row = randint(0, 6)
                    user_ship_col = randint(0, 6)
                    if user_ship_col == 6:
                        user_board2[user_ship_row,user_ship_col-1]='<'
                        user_board2[user_ship_row,user_ship_col]='>'
                    else:
                        user_board2[user_ship_row,user_ship_col]='<'
                        user_board2[user_ship_row,user_ship_col+1]='>'

                else:
                    user_board2[user_ship_row,user_ship_col]='<'
                    user_board2[user_ship_row,user_ship_col+1]='>'
            else:
                if user_board2[user_ship_row,user_ship_col]=='<' or user_board2[user_ship_row,user_ship_col]=='>' or user_board2[user_ship_row,user_ship_col]=='^' or user_board2[user_ship_row,user_ship_col]=='v':
                        
                    user_ship_row = randint(0, 6)
                    user_ship_col = randint(0, 6)
                    user_board2[user_ship_row,user_ship_col-1]='<'
                    user_board2[user_ship_row,user_ship_col]='>'

                else:
                    user_board2[user_ship_row,user_ship_col-1]='<'
                    user_board2[user_ship_row,user_ship_col]='>'
            
            if ai_ship_col==6:

                ai_board2[ai_ship_row,ai_ship_col-1]='<'
                ai_board2[ai_ship_row,ai_ship_col]='>'
                
            elif ai_ship_col==0:
                ai_board2[ai_ship_row,ai_ship_col]='<'
                ai_board2[ai_ship_row,ai_ship_col+1]='>'
                    
            else:
                ai_board2[ai_ship_row,ai_ship_col-1]='<'
                ai_board2[ai_ship_row,ai_ship_col]='>'
                    
                
            
                
                
        else:
            
            #placing random ships
            user_ship_row = randint(0, 6)
            user_ship_col = randint(0, 6)
            ai_ship_row = randint(0, 6)
            ai_ship_col = randint(0, 6)

            
            if user_ship_row == 6:
                if user_board2[user_ship_row,user_ship_col]=='<' or user_board2[user_ship_row,user_ship_col]=='>' or user_board2[user_ship_row,user_ship_col]=='^' or user_board2[user_ship_row,user_ship_col]=='v':
                    user_ship_row = randint(0, 6)
                    user_ship_col = randint(0, 6)
                    user_board2[user_ship_row-1,user_ship_col]='^'
                    user_board2[user_ship_row,user_ship_col]='v'

                else:
                    user_board2[user_ship_row-1,user_ship_col]='^'
                    user_board2[user_ship_row,user_ship_col]='v'
                
            elif user_ship_row == 0:
                if user_board2[user_ship_row,user_ship_col]=='<' or user_board2[user_ship_row,user_ship_col]=='>' or user_board2[user_ship_row,user_ship_col]=='^' or user_board2[user_ship_row,user_ship_col]=='v':
                    user_ship_row = randint(0, 6)
                    user_ship_col = randint(0, 6)
                    if user_ship_row == 6:
                        user_board2[user_ship_row-1,user_ship_col]='^'
                        user_board2[user_ship_row,user_ship_col]='v'
                    else:
                        user_board2[user_ship_row,user_ship_col]='^'
                        user_board2[user_ship_row+1,user_ship_col]='v'

                else:
                    user_board2[user_ship_row,user_ship_col]='^'
                    user_board2[user_ship_row+1,user_ship_col]='v'
            else:
                if user_board2[user_ship_row,user_ship_col]=='<' or user_board2[user_ship_row,user_ship_col]=='>' or user_board2[user_ship_row,user_ship_col]=='^' or user_board2[user_ship_row,user_ship_col]=='v':
                    user_ship_row = randint(0, 6)
                    user_ship_col = randint(0, 6)
                    user_board2[user_ship_row-1,user_ship_col]='^'
                    user_board2[user_ship_row,user_ship_col]='v'

                else:
                    user_board2[user_ship_row-1,user_ship_col]='^'
                    user_board2[user_ship_row,user_ship_col]='v'

            if ai_ship_row==6:
                ai_board2[ai_ship_row-1,ai_ship_col]='^'
                ai_board2[ai_ship_row,ai_ship_col]='v'
               
            elif ai_ship_row==0:
                ai_board2[ai_ship_row,ai_ship_col]='^'
                ai_board2[ai_ship_row+1,ai_ship_col]='v'

            else:
                ai_board2[ai_ship_row-1,ai_ship_col]='^'
                ai_board2[ai_ship_row,ai_ship_col]='v'
        
        
#while play
while play not in dif:
    print('Do you want to play?: Yes or No(1 or 2):')
    play=int(input())



if play == 1:
    
    #setting difficulty
    while level not in dif:
        print('Select difficulty: Easy or Hard(1 or 2):')
        level=int(input())  
        
    #Start random ai game
    if level == 1:    

        #Set ships on the board    
        ship_loc()

        #AI shots with no repetition
        #code from https://stackoverflow.com/questions/40330040/make-dice-values-not-repeat-in-if-statement
        ai_shots_list = []

        dice = random.randint(0,3)


        for i in range(25):
            while True:
                if dice == 0:
                    ai_col = random.randint(0,4)
                    ai_row = random.randint(0,4)
                    ai_shot=(ai_row, ai_col)
                elif dice == 1:
                    ai_col = random.randint(0,4)
                    ai_row = random.randint(0,4)
                    ai_shot=(ai_row, ai_col)
                elif dice == 2:
                    ai_col = random.randint(0,4)
                    ai_row = random.randint(0,4)
                    ai_shot=(ai_row, ai_col)
                else:
                    ai_col = random.randint(0,4)
                    ai_row = random.randint(0,4)
                    ai_shot=(ai_row, ai_col)

                if not ai_shot in ai_shots_list:
                    ai_shots_list.append(ai_shot)
                    break

        for i in ai_shots_list:

            row_shot = int(input("Guess row:"))
            while row_shot not in rowcol:
                print("Invalid value")
                row_shot = int(input("Guess row:"))


            col_shot = int(input("Guess column:"))
            while col_shot not in rowcol:
                print("Invalid value")
                col_shot = int(input("Guess column:"))

            if user_board[row_shot, col_shot] == '*' or user_board[row_shot, col_shot] == 'X': 
                print("We already shot there")


            if user_board[row_shot, col_shot] == 0: 
                print("Admiral, we've missed the enemy battleship!")
                user_board[row_shot,col_shot]= miss
                shot_board[row_shot,col_shot]= miss


            if user_board[row_shot, col_shot] == '<>':
                print("Admiral, we've hit the enemy ship!")
                user_board[row_shot,col_shot]= hit
                shot_board[row_shot,col_shot]= hit
                user_hits_num+=1


            if ai_board[i] == 0: 
                print("Good news! They've missed!")
                ai_board[i]=ai_miss

            if ai_board[i]== '<>':
                print("Attention Admiral! You have been hit!")
                ai_board[i]=ai_hit
                ai_hits_num+=1

            #Display boards to the output
            df_shot_board = pd.DataFrame(shot_board)
            df_ai_board = pd.DataFrame(ai_board)
            display_side_by_side([df_shot_board, df_ai_board], ['Enemy board', 'My board'])


            if user_hits_num==5:
                print("You Won!")
                break
            if ai_hits_num==5:
                print("Bad news! The enemy sank all the ships")
                print("Game Over")
                break
                
                
                
                
                
                
    #Start ai game           
    else:
        #Set ships on the board  
        ship_loc2()
                  
        #AI shots with no repetition
        #code from https://stackoverflow.com/questions/40330040/make-dice-values-not-repeat-in-if-statement
        ai_shots_list = []

        dice = random.randint(0,3)


        for i in range(49):
            while True:
                if dice == 0:
                    ai_col = random.randint(0,6)
                    ai_row = random.randint(0,6)
                    ai_shot=(ai_row, ai_col)
                elif dice == 1:
                    ai_col = random.randint(0,6)
                    ai_row = random.randint(0,6)
                    ai_shot=(ai_row, ai_col)
                elif dice == 2:
                    ai_col = random.randint(0,6)
                    ai_row = random.randint(0,6)
                    ai_shot=(ai_row, ai_col)
                else:
                    ai_col = random.randint(0,6)
                    ai_row = random.randint(0,6)
                    ai_shot=(ai_row, ai_col)

                if not ai_shot in ai_shots_list:
                    ai_shots_list.append(ai_shot)
                    break

        for i in ai_shots_list:

            row_shot = int(input("Guess row:"))
            while row_shot not in rowcol:
                print("Invalid value")
                row_shot = int(input("Guess row:"))


            col_shot = int(input("Guess column:"))
            while col_shot not in rowcol:
                print("Invalid value")
                col_shot = int(input("Guess column:"))

            if user_board2[row_shot, col_shot] == '*' or user_board2[row_shot, col_shot] == 'X': 
                print("We already shot there")


            if user_board2[row_shot, col_shot] == 0: 
                print("Admiral, we've missed the enemy battleship!")
                user_board2[row_shot,col_shot]= miss
                shot_board2[row_shot,col_shot]= miss


            if user_board2[row_shot, col_shot] == '<' or user_board2[row_shot, col_shot] == '>' or user_board2[row_shot, col_shot] == '^' or user_board2[row_shot, col_shot] == 'v':
                print("Admiral, we've hit the enemy ship!")
                user_board2[row_shot,col_shot]= hit
                shot_board2[row_shot,col_shot]= hit
                user_hits_num+=1


            if ai_board2[i] == 0: 
                print("Good news! They've missed!")
                ai_board2[i]=ai_miss
                
                                
            if shot_near!=1:
                if ai_board2[i]== '<' or ai_board2[i]== '>' or ai_board2[i]== '^' or ai_board2[i]== 'v':
                    print("Attention Admiral! You have been hit!")
                    ai_board2[i]=ai_hit
                    ai_hits_num+=1
                    past_shot=i
                    shot_near=1 
                    
            else:
                for i in ai_shots_list:
                    ai_row = past_shot[0]
                    ai_col = past_shot[1]

                    if ai_row == 6 or ai_col == 6:
                        if ai_board2[ai_row-1,ai_col]== '<' or ai_board2[ai_row-1,ai_col]== '>' or ai_board2[ai_row-1,ai_col]== '^' or ai_board2[ai_row-1,ai_col]== 'v':
                            print("Attention Admiral! You have been hit!")
                            ai_board2[ai_row-1,ai_col]=ai_hit
                            ai_hits_num+=1
                            shot_near=0
                            break
                        elif ai_board2[ai_row-1,ai_col] == 0: 
                            print("Good news! They've missed!")
                            ai_board2[ai_row-1,ai_col]=ai_miss

                        else:
                            print("Attention Admiral! You have been hit!")
                            ai_board2[ai_row,ai_col-1]=ai_hit
                            ai_hits_num+=1
                            shot_near=0
                            break
                        
                        
                    elif ai_row == 0 or ai_col == 0:
                        if ai_board2[ai_row+1,ai_col]== '<' or ai_board2[ai_row+1,ai_col]== '>' or ai_board2[ai_row+1,ai_col]== '^' or ai_board2[ai_row+1,ai_col]== 'v':
                            print("Attention Admiral! You have been hit!")
                            ai_board2[ai_row+1,ai_col]=ai_hit
                            ai_hits_num+=1
                            shot_near=0
                            break
                        elif ai_board2[ai_row+1,ai_col] == 0: 
                            print("Good news! They've missed!")
                            ai_board2[ai_row+1,ai_col]=ai_miss

                        elif ai_board2[ai_row,ai_col+1]== '<' or ai_board2[ai_row,ai_col+1]== '>' or ai_board2[ai_row,ai_col+1]== '^' or ai_board2[ai_row,ai_col+1]== 'v':
                            print("Attention Admiral! You have been hit!")
                            ai_board2[ai_row,ai_col+1]=ai_hit
                            ai_hits_num+=1
                            shot_near=0
                            break
                        elif ai_board2[ai_row,ai_col+1] == 0: 
                            print("Good news! They've missed!")
                            ai_board2[ai_row,ai_col+1]=ai_miss
                            
                            
                        
                        
                    else:
                    
                        if ai_board2[ai_row-1,ai_col]== '<' or ai_board2[ai_row-1,ai_col]== '>' or ai_board2[ai_row-1,ai_col]== '^' or ai_board2[ai_row-1,ai_col]== 'v':
                            print("Attention Admiral! You have been hit!")
                            ai_board2[ai_row-1,ai_col]=ai_hit
                            ai_hits_num+=1
                            shot_near=0
                            break
                        elif ai_board2[ai_row-1,ai_col] == 0: 
                            print("Good news! They've missed!")
                            ai_board2[ai_row-1,ai_col]=ai_miss
                            break

                        elif ai_board2[ai_row+1,ai_col]== '<' or ai_board2[ai_row+1,ai_col]== '>' or ai_board2[ai_row+1,ai_col]== '^' or ai_board2[ai_row+1,ai_col]== 'v':
                            print("Attention Admiral! You have been hit!")
                            ai_board2[ai_row+1,ai_col]=ai_hit
                            ai_hits_num+=1
                            shot_near=0
                            break
                        elif ai_board2[ai_row+1,ai_col] == 0: 
                            print("Good news! They've missed!")
                            ai_board2[ai_row+1,ai_col]=ai_miss
                            

                        elif ai_board2[ai_row,ai_col-1]== '<' or ai_board2[ai_row,ai_col-1]== '>' or ai_board2[ai_row,ai_col-1]== '^' or ai_board2[ai_row,ai_col-1]== 'v':
                            print("Attention Admiral! You have been hit!")
                            ai_board2[ai_row,ai_col-1]=ai_hit
                            ai_hits_num+=1
                            shot_near=0
                            break
                        elif ai_board2[ai_row,ai_col-1] == 0: 
                            print("Good news! They've missed!")
                            ai_board2[ai_row,ai_col-1]=ai_miss

                        elif ai_board2[ai_row,ai_col+1]== '<' or ai_board2[ai_row,ai_col+1]== '>' or ai_board2[ai_row,ai_col+1]== '^' or ai_board2[ai_row,ai_col+1]== 'v':
                            print("Attention Admiral! You have been hit!")
                            ai_board2[ai_row,ai_col+1]=ai_hit
                            ai_hits_num+=1
                            shot_near=0
                            break
                        elif ai_board2[ai_row,ai_col+1] == 0: 
                            print("Good news! They've missed!")
                            ai_board2[ai_row,ai_col+1]=ai_miss
                                    
                    
                

            #Display boards to the output
            df_shot_board2 = pd.DataFrame(shot_board2)
            df_ai_board2 = pd.DataFrame(ai_board2)
            display_side_by_side([df_shot_board2, df_ai_board2], ['Enemy board', 'My board'])


            if user_hits_num==9:
                print("You Won!")
                break
            if ai_hits_num==9:
                print("Bad news! The enemy sank all the ships")
                print("Game Over")
                break

else:
    print("Thanks for playing")
```
